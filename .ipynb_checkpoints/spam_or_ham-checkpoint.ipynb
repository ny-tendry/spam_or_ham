##### etape du NLP
- nettoyage basique (miniscule, ponctuation, url)
- tokenisation
- suppression des stopwords
- lemmatisation
- bag of words (vectorisation des messages)

  


```python
def sms_cleanner(message):
    message = message.lower() # on transforme en minuscule tous les lettres
    message = re.sub(r"['|-]",' ', message) # on remplace [' | -] par les espaces
    message = re.sub(r'http\S+|www\S+', '<url>', message) # on remplace tous les url par un token <url>
    message = re.sub(r'\w+.@\w+.\w+', 'email', message) # on remplace aussi les email par <email>
    message = re.sub(r'\+\s?\d+', '<numero mobile>', message) # on remplace aussi les numeros par <phone_number>
    message = re.sub(r'\d+', '<chiffre>', message) # on remplace les chiffre par <number>
    message = re.sub(r'[^\w\s]', '', message) # on supprime les ponctuations

    list_words = message.split()

    stop_words = stopwords.words('french')
    list_words = [word for word in list_words if word not in stop_words]

    return ' '.join(list_words)

df_message["text_fr_cleaned"] = df_message["text_fr"].apply(sms_cleanner) 

nlp = spacy.load("fr_core_news_sm") # documentation https://www.codeandcortex.fr/spacy-natural-langage-process-lemmatisation/

def lemmatisation(message):
    list_lemma = []
    doc = nlp(message)
    for token in doc:
        list_lemma.append(token.lemma_)
    return ' '.join(list_lemma)

df_message["text_fr_cleaned"] = df_message["text_fr_cleaned"].apply(lemmatisation)

vec = CountVectorizer(max_features=5000, ngram_range=(1, 2))

X = vec.fit_transform(df_message['text_fr_cleaned']).toarray()
```

```python
#encodage du feature

encoder = LabelEncoder()

y = encoder.fit_transform(df_message["labels"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
cv = StratifiedKFold(n_splits=5,shuffle=True, random_state=42)
model1 = LogisticRegression()
```

```python
scores = cross_val_score(model1, X_train, y_train, cv=cv)
print(f"Estimation de la performance future: {scores.mean():.2f}")
```

```python
model1.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model1.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```
